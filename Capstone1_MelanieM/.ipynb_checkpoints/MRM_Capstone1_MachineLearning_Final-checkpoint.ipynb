```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
import itertools
import seaborn as sns
```

```python
sfo_data = pd.read_csv('2017_SFO_Customer_Survey.csv')
```

I am taking the position of an airline marketing an airline miles credit card. Because business travelers are a key demographic for airlines, targeting business travelers for an airline miles credit card is a smart business decision. I have often seen airlines set up booths inside the terminal to market their credit cards. However, I believe based on the statistical analysis that I have done that it might be possible to additionally target customers based on whether or not they checked a bag, which is something that you would know when a customer checked in at the airport. Additionally, an airline would have access to records of how often that person took that airline in the last year. I therefore believe that I should try and predict whether someone is a business or non-business traveler based on whether or not they checked a bag, their age, their gender, and the number of times flown in the past year, which are all variables in the survey. They could then target business traveler with an offer when they checked in to the flight.

```python
sfo_data = sfo_data.rename(columns={'Q2PURP1':'Purpose', 'Q3GETTO1':'Transportation', 'Q4BAGS':'Bags', 
                        'Q4STORE':'Store', 'Q4FOOD':'Food', 'Q4WIFI':'WiFi', 'Q5TIMESFLOWN':'Times Flown',
                        'Q11TSAPRE':'TSA Pre-Check', 'Q19AGE':'Age', 'Q20GENDER':'Gender', 'Q21INCOME':'Income'})
```

```python
sfo_data['Purpose'] = sfo_data['Purpose'].map({1:'Business', 2:'Pleasure', 3:'Visit others', 4:'School', \
                                              5:'Conference', 6:'Life event', 7:'Other', \
                                               10:'Escorting/personal errands/medical', 11:'Military', \
                                              12:'Volunteer/political/religious', 13:'Moving/immigration', \
                                              0:'Blank/non-response'})
```

```python
sfo_data['Bags'] = sfo_data['Bags'].map({1:'Yes', 2:'No', 3:"Don't know", 0:'Blank/multiple responses'})
```

```python
sfo_data['Age'] = sfo_data['Age'].map({1:'Under18', 2:'18-24', 3:'25-34', 4:'35-44', 5:'45-54', 6:'55-64', \
                                            7:"65Over", 8:"Don't know/Refused", 0:'Blank/Multiple Responses'})
```

```python
sfo_data['Gender'] = sfo_data['Gender'].map({1:'Male', 2:'Female', 3:'Other', 0:'Blank/Multiple responses'})
```

```python
sfo_data['Times Flown'] = sfo_data['Times Flown'].map({1:'1', 2:'2', 3:'3-6', 4:'7-12', 5:'13-24', 6:'>24', \
                                                         0:'Blank/Multiple Responses'})
```

```python
sfo_data_2 = sfo_data[sfo_data['Purpose'] != 'Blank/non-response'] #drop all rows that did not specify purpose
```

```python
sfo_data_2['Business?'] = sfo_data_2['Purpose'].apply(lambda x: 
                                            1 if x == 'Business'
                                            else 0)
```

```python
relevant_data = sfo_data_2[['Business?','Bags','Age','Gender','Times Flown','WEIGHT']]
```

```python
X_sfodata = relevant_data[['Bags','Age','Gender','Times Flown']]
y_sfodata = relevant_data['Business?']
weights_sfodata = relevant_data['WEIGHT']
```

```python
#encoding to dummy variables
sfo_data_dummies = pd.get_dummies(X_sfodata)
sfo_data_dummies.columns
```

```python
sfo_data_dropped = sfo_data_dummies.drop(['Bags_Blank/multiple responses',
                                          "Bags_Don't know",'Age_Blank/Multiple Responses',
                      "Age_Don't know/Refused",'Gender_Blank/Multiple responses',
                       'Times Flown_Blank/Multiple Responses'],axis=1)
```

```python
sfo_data_dropped.head()
```

```python
sfo_data_dropped.columns
```

Logistic regression with no class balancing:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import recall_score

c_space = np.logspace(-5,8,15)
param_grid = {'C':c_space, 'penalty':['l1','l2']}

X_train, X_test, y_train, y_test, sw_train, sw_test = \
train_test_split(sfo_data_dropped, y_sfodata, weights_sfodata, test_size=0.3, random_state=42)

logreg_simple = LogisticRegression()

logreg_cv_simple = GridSearchCV(logreg_simple, param_grid, cv=5, scoring='roc_auc')
logreg_cv_simple.fit(X_train,y_train,sample_weight=sw_train)
```

```python
logreg_cv_simple.best_params_
```

```python
pred_logreg_cv_simple_proba = logreg_cv_simple.predict_proba(X_test)
```

```python
pred_logreg_cv_simple_proba_list = pred_logreg_cv_simple_proba[:,1].tolist()
```

```python
#function to calculate max f1 score

from sklearn.metrics import f1_score

def max_f1_score(gridsearch,data_test,pred_test,sampleweight_test):
    pred_gridsearch_proba = gridsearch.predict_proba(data_test)
    pred_gridsearch_proba_list = pred_gridsearch_proba[:,1].tolist()
    f1_score_max = 0
    y_pred_max = np.empty(len(pred_test),)
    for probability in pred_gridsearch_proba_list:
        y_pred = (gridsearch.predict_proba(data_test)[:,1] >= probability).astype(bool)
        f1_score_current = f1_score(pred_test, y_pred, sample_weight = sampleweight_test, pos_label=1)
        if f1_score_current > f1_score_max:
            f1_score_max = f1_score_current
            y_pred_max = y_pred
    return f1_score_max, y_pred_max
```

```python
#max f1 score for simple logreg cv
f1_score_max_logreg_cv_simple, y_pred1 = max_f1_score(logreg_cv_simple,X_test,y_test,sw_test)
print(f1_score_max_logreg_cv_simple)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1, sample_weight = sw_test))
```

Having tried logistic regression, random forests, and SVM, and seeing that logistic regression works the best, I will now try doing logistic regression with upsampling instead of class_balance.

```python
from sklearn.utils import resample

# concatenate our training data back together
X_resample = pd.concat([X_train, y_train, sw_train], axis=1)

# separate minority and majority classes
not_business = X_resample[X_resample['Business?']==0]
business = X_resample[X_resample['Business?']==1]

# upsample minority
business_upsampled = resample(business,
                          replace=True, # sample with replacement
                          n_samples=len(not_business), # match number in majority class
                          random_state=42) # reproducible results

upsampled = pd.concat([not_business, business_upsampled])
upsampled.head()
```

```python
X_train_upsampled = upsampled.drop('Business?', axis=1)
X_train_upsampled = X_train_upsampled.drop('WEIGHT', axis=1)
X_train_upsampled.head()
```

```python
y_train_upsampled = upsampled['Business?']
sw_train_upsampled = upsampled['WEIGHT']
```

```python
logreg3 = LogisticRegression()

logreg_cv_3 = GridSearchCV(logreg3, param_grid, cv=5, scoring='roc_auc')
logreg_cv_3.fit(X_train_upsampled,y_train_upsampled,sample_weight=sw_train_upsampled)
```

```python
logreg_cv_3.best_params_
```

```python
f1_score_max_logreg_cv_3, y_pred7 = max_f1_score(logreg_cv_3,X_test,y_test,sw_test)
print(f1_score_max_logreg_cv_3)
```

```python
print(classification_report(y_test, y_pred7, sample_weight = sw_test))
```
