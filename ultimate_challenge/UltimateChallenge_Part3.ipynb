The last trip date is July 1st, 2014, so I will say that the data was pulled on July 1st. June is 30 days long, so I will consider users retained if they took a trip between June 1st and July 1st.

```python
from datetime import datetime as dt
datetime_june1st = dt.strptime('June 1 2014', '%B %d %Y')

df['retained'] = (df['last_trip_date'] >= datetime_june1st)
```

For predictive model, I want to predict whether user was retained (had a June trip) based on the other columns in the model. I will not include the last_trip_date column since I only want to predict whether someone had a June trip or not. However, I will include signup_date because it is possible that users who signed up later in January might be more likely to have a trip in June. The other columns should all be predictive as well.

I will have to convert the city and phone columns to numeric oolumns using pd.get_dummies.

```python
df_no_last_trip_date = df.drop('last_trip_date', axis=1)
```

```python
df_no_last_trip_date
```

```python
df_no_last_trip_date['signup_date'] = pd.to_datetime(df_no_last_trip_date['signup_date'])
df_no_last_trip_date['signup_day'] = pd.DatetimeIndex(df_no_last_trip_date['signup_date']).day
df_no_last_trip_date
```

```python
df_no_last_trip_date['ultimate_black_user'] = df_no_last_trip_date['ultimate_black_user'].astype(int)
```

```python
df_no_last_trip_date['retained'] = df_no_last_trip_date['retained'].astype(int)
```

```python
df_no_last_trip_date_2 = df_no_last_trip_date.drop('signup_date', axis=1)
df_no_last_trip_date_2
```

```python
X = df_no_last_trip_date_2.drop('retained', axis=1)
y = df_no_last_trip_date_2['retained']
```

```python
X
```

```python
y
```

```python
X_dummies = pd.get_dummies(X, drop_first=True)
X_dummies.columns
```

Since the data is unbalanced, it might be useful to use the class balanced parameter. I will try logistic regression, random forests, and SVM.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import recall_score

c_space = np.logspace(-5,8,15)
param_grid = {'C':c_space, 'penalty':['l1','l2']}

X_train, X_test, y_train, y_test= train_test_split(X_dummies, y, test_size=0.3, random_state=42)

logreg_classbalanced = LogisticRegression(class_weight='balanced')

logreg_cv_class_balanced = GridSearchCV(logreg_classbalanced, param_grid, cv=5, scoring='f1')
logreg_cv_class_balanced.fit(X_train, y_train)
```

```python
y_pred1 = logreg_cv_class_balanced.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```

Trying random forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

param_grid_rfc = { 
    'n_estimators': [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
    'max_features': ['auto', 'sqrt', 'log2']
}

rfc = RandomForestClassifier(random_state=42, class_weight='balanced')
rfc_cv = GridSearchCV(rfc, param_grid_rfc, cv=5, scoring='f1', verbose=2)
rfc_cv.fit(X_train,y_train)
```

```python
y_pred2 = rfc_cv.predict(X_test)
```

```python
print(classification_report(y_test, y_pred2))
```

I will create a decision tree so I can visualize the data.

```python
from sklearn.tree import DecisionTreeClassifier
dtree1 = DecisionTreeClassifier(max_depth=3, random_state=42, class_weight='balanced')
dtree1.fit(X_train,y_train)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree1, out_file=dot_data, filled=True,
                feature_names = X_train.columns, class_names=['Not retained', 'Retained'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

It looks like the average rating by the driver and whether the city was King's Landing are the two most important factors. If the average rating by the driver was less than 4.05 then that group of riders was not retained. Also, if the city is King's Landing, then that group of riders was retained, and the riders who were not from King's Landing were not retained as much.
