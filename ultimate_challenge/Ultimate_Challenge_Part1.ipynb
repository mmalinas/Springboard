```python
#import json
import json
import pandas as pd

f = open('logins.json')
data = json.load(f)
df = pd.DataFrame.from_dict(data, orient='columns')
df
```

```python
df.tail()
```

It looks like the data ranges from January 1st, 1970 to April 13th, 1970 and has 93,142 data points.

```python
df['login_time'].isnull().values.any()
```

There are no missing values. However, it looks like the timestamps are out of order, which may be a problem. I will see if I can put them in order.

```python
df['count'] = 1
df['login_time'] = pd.to_datetime(df['login_time'])
df_sorted = df.sort_values(by='login_time')
df_sorted
```

```python
df_sums = df_sorted.set_index('login_time').resample('15T').sum()
df_sums
```

```python
import matplotlib.pyplot as plt
df_sums.plot(figsize=(15,4))
plt.show()
```

This is a plot of the data in 15 minute intervals. I will now try and resample the data to look at it over 1 day intervals which might give me more insight.

```python
df_oneday = df_sorted.set_index('login_time').resample('1D').sum()
df_oneday.plot(figsize=(15,4))
plt.show()
```

It looks like the average count of login times goes up between January and April, but there are fluctuations between days, with about 4-5 peaks per month.

```python
df_oneday
```

I will now look at the first five days of the dataframe, sampled at 1 hour intervals, to give me some idea of the daily fluctuations.

```python
df_twohours = df_sorted.set_index('login_time').resample('2H').sum()
df_twohours_firstfivedays = df_twohours.first('5D')
df_twohours_firstfivedays
```

```python
df_twohours_firstfivedays.plot(figsize=(15,4))
plt.show()
```

It looks like there are some fluctuations over the course of a day, but not necessarily in a predictable way.
